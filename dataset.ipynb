```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("AmesHousing.csv")

# Preview
df.head()
```

```python
# Shape of dataset
print("Shape:", df.shape)

# Data types
df.info()

# Summary statistics
df.describe()
```

```python
# Count missing values per column
missing = df.isnull().sum()

# Show only columns with missing values
missing[missing > 0].sort_values(ascending=False)
```

```python
from sklearn.preprocessing import LabelEncoder

df_encoded = df.copy()

for col in df_encoded.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df_encoded[col] = df_encoded[col].astype(str)
    df_encoded[col] = le.fit_transform(df_encoded[col])

df_encoded.head()
```

```python
# Convert to numpy
data = df_encoded.to_numpy()

# Flatten
flat = data.flatten()

# Resize to 900 elements
if len(flat) < 900:
    flat = np.pad(flat, (0, 900 - len(flat)), mode='constant')
else:
    flat = flat[:900]

matrix_30x30_raw = flat.reshape(30, 30)

print("Shape:", matrix_30x30_raw.shape)
matrix_30x30_raw
```

```python
from sklearn.impute import SimpleImputer

df_fixed = df.copy()

# Separate types
num_cols = df_fixed.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df_fixed.select_dtypes(include=['object']).columns

# Numeric → median
num_imputer = SimpleImputer(strategy='median')
df_fixed[num_cols] = num_imputer.fit_transform(df_fixed[num_cols])

# Categorical → most frequent
cat_imputer = SimpleImputer(strategy='most_frequent')
df_fixed[cat_cols] = cat_imputer.fit_transform(df_fixed[cat_cols])
```

```python
df_fixed_encoded = df_fixed.copy()

for col in df_fixed_encoded.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df_fixed_encoded[col] = le.fit_transform(df_fixed_encoded[col].astype(str))
```

```python
data_clean = df_fixed_encoded.to_numpy()

flat_clean = data_clean.flatten()

if len(flat_clean) < 900:
    flat_clean = np.pad(flat_clean, (0, 900 - len(flat_clean)), mode='constant')
else:
    flat_clean = flat_clean[:900]

matrix_30x30_clean = flat_clean.reshape(30, 30)

print("Final Shape:", matrix_30x30_clean.shape)
matrix_30x30_clean
```
